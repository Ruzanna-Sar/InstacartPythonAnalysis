## Step 4 - Creating loyalty flag

```python
ords_prods_merge['max_order']=ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#checking the creation of a new column called 'max_order'
ords_prods_merge.head(2)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.head(2)
```

## Step 5- Statistics for loyalty category

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'max', 'min']})
```

On average Loyal customes pay the least whereas the new customer pays the most. maybe loyal customers buy more items
and on average their product portfolio has the lowest average price

## Step 6 - Creating a Spending Flag

```python
ords_prods_merge['average_price']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(2)
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.head(5)
```

## Step 7- Creating Frequency Flag

```python
ords_prods_merge['median_order']=ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order'] > 20, 'Frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order'] > 10) & (ords_prods_merge['median_order'] <= 20), 'Frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order'] <= 10, 'Frequency_flag'] = 'Frequent customer'
```
